```python
n_p,n_d,p_dim=input_map.size()
self=net
id_map=0
verbose=0
X=None
if True:
        p_shape=self.n_patches[id_map]                    ## p_shape: dimension of the new patches


        #### STEP 1: normalize data
        if X==None:
            input_map=self.Kernel[id_map].all_patches         ## input map: should check that it is of dim
        else:
            input_map=X
        if len(input_map.size())==4:                      ## [nb_patches x nb_graphs x dim_patch] e.g, in MNIST there are 28 x28 patches,  so for a filter of size 5, this is of dimension  784 x60000 x 25
            input_map=input_map.view(input_map.size()[0],input_map.size()[1],-1).contiguous()
        n_p,n_d,p_dim=input_map.size()
        print(input_map.size())
        if self.Kernel[id_map].pca!=None:
            input_map=self.Kernel[id_map].pca.transform(input_map.contiguous().view(n_p*n_d,p_dim).numpy())
            input_map=torch.Tensor(input_map).view(n_p,n_d,-1)
            print(input_map.size())
            n_p,n_d,p_dim=input_map.size()
        
        #### STEP 2: extract relevant parameters 
        spacing=2#np.min([input_map.size()[0]//p_shape,input_map.size()[1]//p_shape])  
        gamma=self.subsampling_factors[id_map]
        beta=gamma*spacing
        batch_size=self.batch_size[id_map]
        D=self.Kernel[id_map].n_components
        sigma=self.Kernel[id_map].sigma
        
        ### STEP 3: Compute the normalized version of the patchs
        norm = input_map.norm(p=2, dim=2, keepdim=True)      # [nb_patches x nb_graphs ] 
        if verbose: print('dim norm', norm.size())
        if verbose: print('max norm',torch.max(norm))
        if verbose: print('min norm',torch.min(norm[norm>0]))
        norm2=norm.clone()
        epsilon=0.0001
        norm2[norm2<epsilon]=1                                ### Capping of th elower bound for the norm  (to tackle the case of the patches with variance 0)
        norm_patches=torch.div(input_map,norm2.expand_as(input_map))
        N=norm_patches.size()[1]   
        
        
        ## STEP 4: Extract patches with the size of the current image  (dim)
        dim=int(np.sqrt(input_map.size()[0]))     ###  this is the size of the new patch
        size_patch=p_shape
        mpatches=extract_patch_mask(batch_size,[dim,dim],size_patch, beta=beta,zero_padding=True)
        mpatches=mpatches.view(-1,dim*dim)
        selected_pixels=[i*dim+j for j in np.arange(0,dim, spacing) for i in np.arange(0,dim, spacing)]
        if verbose: print(selected_pixels)
            
            
        ### STEP 5: Compute activation set in a batch-wise fashion
        output_map=torch.Tensor(len(selected_pixels),n_d,D)
        for b in range(N//batch_size):
            print(b)
            temp=torch.Tensor(D,n_p*batch_size)
            zeta=torch.Tensor(D,n_p*batch_size)
            #print(norm_patches.size())
            XX=norm_patches[:,b*batch_size:(b+1)*batch_size,:].contiguous().view(-1,p_dim)
            sXX=torch.sum((XX)**2,dim=1)
            sWW=torch.sum((self.Kernel[id_map].W.view(D,p_dim))**2,dim=1)#.expand_as(XX)   ## XX.size()=WW.size()=[D,n_p*batch_size,p_dim ] => sum((XX-WW)**2,2).size()=[D,n_p*batch_size]
            sXW=torch.matmul(XX,self.Kernel[id_map].W.t().data)
            print('XX',XX.size())
            print('sXX',sXX.size())
            print('sWW',sWW.size())
            print('sXW',sXW.size())
            tot=sXX.view(-1,1).expand_as(sXW)*(sXW*sWW.data.view(1,-1).expand_as(sXW))
            print('tot',tot.size())
            w=torch.sqrt(self.Kernel[id_map].eta.view(1,D).data.expand_as(tot))*torch.exp(torch.div(tot,-sigma))  ### temporary: zeta=  eta_L e^(-1/sigma_k**2 *||Psi_tilde(z)-W_L||^2)
            temp=w
            print(output_map)
            print('nb 1 of null is ',torch.sum(temp!=temp))
            #print(w.size())
            Reg=norm[:,b*batch_size:(b+1)*batch_size].squeeze().contiguous().view(1,n_p*batch_size).expand_as(w)
            #print(Reg.size())
            zeta=Reg*temp#output2=output.view(n_d*n_p,D,p_dim)    ### zeta=  ||Psi(z)||_2 eta_L e^(-1/sigma_k**2 *||Psi_tilde(z)-W_L||^2)
            ## zeta has how size: [D,n_p*batch_size] and Reg:[n_p*batch_size,1]
            print('nb 2 of null is ',torch.sum(zeta!=zeta))
            zeta=zeta.view(D,n_p,batch_size)      ### gather tensor
            zeta2=zeta.permute(1,2,0)
            test=torch.matmul(mpatches,zeta2.contiguous().view(zeta2.size()[0],zeta2.size()[1]*zeta2.size()[2]))
            out=test.view(zeta2.size()[0],zeta2.size()[1],zeta2.size()[2])
            print('nb 3 of null is ',torch.sum(out!=out))
            #print(out)
            output_map[:,b*batch_size:(b+1)*batch_size,:]=out[selected_pixels,:,:]   
        print('nb final of null is ',torch.sum(output_map!=output_map))   
        output_map_norm=output_map #normalize_output(output_map, verbose=False)
        self.Kernel[id_map].output=output_map_norm
```

```python
def extract_patches_from_vector(X,size_patch, zero_padding=True):

    nb_locations=X.size()[0]
    x_dim=int(np.sqrt(X.size()[0]))
    nb_data_points=X.size()[1]
    D=X.size()[2]
    
    #####
    size_image=[x_dim,x_dim]
    a=2*np.max([size_patch//2,1])+size_image[0]
    b=2*np.max([size_patch//2,1])+size_image[1]
    padded_image=np.zeros((a,b))
    padded_image[np.max([size_patch//2,1]):np.max([size_patch//2,1])+size_image[0],np.max([size_patch//2,1]):(np.max([size_patch//2,1]))+size_image[1]]=1
    nx,ny=padded_image.shape
    test=torch.Tensor([create_BW_mask(size_image,size_patch,ii,jj)[np.max([size_patch//2,1]):np.max([size_patch//2,1])+size_image[0],np.max([size_patch//2,1]):(np.max([size_patch//2,1]))+size_image[1]] \
        for ii in np.arange(0,nx-2*np.max([size_patch//2,1])) \
        for jj in np.arange(0,ny-2*np.max([size_patch//2,1])) ])
    test=test.view(test.size()[0],-1)
    #for i in xrange(0,batch_size*(data.size()[1]//batch_size),batch_size):
    test2=torch.stack([torch.cat([X[u,:,:].t() for u in np.where(test[j,:].numpy()!=0)[0]]\
                +[torch.Tensor(D,nb_data_points)]*(size_patch**2-len(np.where(test[j,:].numpy()!=0)[0]))) for j in range(nb_locations)])
    
    return test2
```

```python
def extract_patches_from_image(data,size_patch, zero_padding=True):
	size_image=data.size()[1:]
	a=size_patch+size_image[0]
	b=size_patch+size_image[1]
	padded_image=torch.Tensor(np.zeros((data.size()[0],a,b)))
	padded_image[:,size_patch//2:(size_patch//2)+size_image[0],size_patch//2:(size_patch//2)+size_image[1]]=data
	nx,ny=padded_image.size()[1:]
	patches=torch.Tensor(np.array([padded_image[:,ii:ii+size_patch,jj:jj+size_patch].numpy() for ii in np.arange(0,nx-size_patch,1) for jj in np.arange(0,nx-size_patch,1) ]))
	return patches
```

```python
def extract_patches_from_vector(data,size_patch,beta=1, zero_padding=True):
    nb_locations=data.size()[0]
    x_dim=int(np.sqrt(data.size()[0]))
    nb_data_points=data.size()[1]
    D=data.size()[2]
    test=extract_patch_mask(nb_data_points,[x_dim,x_dim],size_patch, beta=beta,zero_padding=True)
    test=test.view(test.size()[0],-1)
    test2=torch.stack([torch.cat([data[u,:,:].t() for u in np.where(test[j,:].numpy()!=0)[0]]\
                +[torch.Tensor(D,nb_data_points)]*(x_dim**2-len(np.where(test[j,:].numpy()!=0)[0]))) for j in range(nb_locations)])
    return test2
```
